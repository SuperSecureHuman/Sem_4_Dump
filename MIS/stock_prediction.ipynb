```python
import pandas as pd
```

```python
df = pd.read_csv('AAPL.csv')
```

```python
# drop these columns
df.drop(["Date",'Date_col', 'Day', 'DayofWeek', 'DayofYear', 'Week', 'Is_month_end', 'Is_month_start', 'Is_quarter_end', 'Is_quarter_start', 'Is_year_end', 'Is_year_start', 'Is_leap_year', 'Year', 'Month'], axis=1, inplace=True)
```

```python
df
```

```python
# Last columsn is the target
y = df.iloc[:, -1].values
# drop the last column
X = df.iloc[:, :-1].values

# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

# 80% training and 20% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
# Scale the data

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

```

```python
from torch_gpu import ADMM

admm = ADMM(X_train, y_train)
```

```python
from tqdm.notebook import tqdm
```

```python
for i in tqdm(range(100)):
    admm.step()
```
