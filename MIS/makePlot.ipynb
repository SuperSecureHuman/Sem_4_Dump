```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("benchmark_results.csv")
```

```python
df
```

```python
plt.figure(figsize=(25, 15))

# Plot admm_iter_time with smoothing
sns.lineplot(x=df["n_samples"], y=df["admm_iter_time"], label="Parallel", linewidth=2.5, ci=None)

# Plot admm_parallel_time with smoothing
sns.lineplot(x=df["n_samples"], y=df["admm_parallel_time"], label="Iterative", linewidth=2.5, ci=None)

# Plot admm_cpu_time with smoothing
sns.lineplot(x=df["n_samples"], y=df["admm_cpu_time"], label="Torch CPU", linewidth=2.5, ci=None)

# Plot admm_gpu_time with smoothing
sns.lineplot(x=df["n_samples"], y=df["admm_gpu_time"], label="Torch GPU", linewidth=2.5, ci=None)

plt.xlabel("n_samples")
plt.ylabel("Time (seconds)")
plt.title("ADMM Time Comparison")
plt.legend()
plt.show()
```
